```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('/content/apple_quality.csv')
data
```

```python
data.describe()
```

```python
data.corr()
```

```python
df = pd.DataFrame(data)
```

```python
df.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
x = data[['Size','Weight','Sweetness','Crunchiness','Juiciness','Ripeness','Acidity']]
y = data['Quality']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.10)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
# Training three SVC models for the same data to compare the models

from sklearn.svm import SVC
# Creating the model
model_1 = SVC(kernel = 'rbf')
model_2 = SVC(kernel = 'sigmoid')
model_3 = SVC(kernel = 'linear')
# Training the model
model_1.fit(x,y)
model_2.fit(x,y)
model_3.fit(x,y)

y1_pred = model_1.predict(x_test)
y2_pred = model_2.predict(x_test)
y3_pred = model_3.predict(x_test)
```

```python
y1_pred = pd.Series(y1_pred)
y2_pred = pd.Series(y2_pred)
y3_pred = pd.Series(y3_pred)

print(y1_pred)
print(y2_pred)
print(y3_pred)
```

```python
from sklearn import metrics
```

```python
accuracy_1 = metrics.accuracy_score(y_test,y1_pred)
accuracy_2 = metrics.accuracy_score(y_test,y2_pred)
accuracy_3 = metrics.accuracy_score(y_test,y3_pred)

confusion_matrix_1 = metrics.confusion_matrix(y_test,y1_pred)
confusion_matrix_2 = metrics.confusion_matrix(y_test,y2_pred)
confusion_matrix_3 = metrics.confusion_matrix(y_test,y3_pred)

classification_report_1 = metrics.classification_report(y_test,y1_pred)
classification_report_2 = metrics.classification_report(y_test,y2_pred)
classification_report_3 = metrics.classification_report(y_test,y3_pred)
```

```python
print(f'Accuracy_1 = {accuracy_1}')
print(f'Accuracy_2 = {accuracy_2}')
print(f'Accuracy_3 = {accuracy_3}')

print(f'Confusion_matrix_1 = {confusion_matrix_1}')
print(f'Confusion_matrix_2 = {confusion_matrix_2}')
print(f'Confusion_matrix_3 = {confusion_matrix_3}')

print(f'Classification_report_1 = {classification_report_1}')
print(f'Classification_report_2 = {classification_report_2}')
print(f'Classification_report_3 = {classification_report_3}')
```

```python
# Plotting a bar graph to see which model is best out of three based on visual interpretation
scores = [accuracy_1,accuracy_2,accuracy_3]
bins = ['model_1','model_2','model_3']
colour = ['red','green','orange']
labels = ['rbf','sigmoid','linear']

plt.bar(bins,scores,color=colour,label=labels)

plt.xlabel("Models")
plt.ylabel("Accuracy Score")
plt.title("Comparison of Accuracy scores")
plt.legend()
```
